```
import pandas as pd # データフレームワーク処理のライブラリをインポート
#df = pd.read_csv("data.csv") # データの読み込み
df = pd.read_csv("/content/sample_data/6-8.csv") # データの読み込み
```

```
df.head() #データの確認

print(df.head(10))
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler([-1,1])

df1 = df.drop('id', axis = 1)

scaler.fit(df1)

data_MinMaxScaler = scaler.transform(df1)
```

```
df = pd.DataFrame(data_MinMaxScaler)
```

```
print(df)
```

```
df = df.reset_index()
df
```

```
# 図やグラフを図示するためのライブラリをインポートする。
import matplotlib.pyplot as plt
%matplotlib inline
from pandas import plotting # 高度なプロットを行うツールのインポート
plotting.scatter_matrix(df, figsize=(6,6), alpha=0.8, diagonal='kde')   #全体像を眺める
plt.show()
```

```
from sklearn.cluster import KMeans # K-means クラスタリングをおこなう
# この例では 3 つのグループに分割 (メルセンヌツイスターの乱数の種を 10 とする)
kmeans_model = KMeans(n_clusters=4, random_state=10).fit(df.iloc[:, 1:])
```

```
# 分類結果のラベルを取得する
labels = kmeans_model.labels_
```

```
# 分類結果を確認
labels
```

```
# それぞれに与える色を決める。
color_codes = {0:'#00FF00', 1:'#FF0000', 2:'#0000FF', 3:'#000000', }
```

```
# サンプル毎に色を与える。
colors = [color_codes[x] for x in labels]
```

```
# 色分けした Scatter Matrix を描く。
plotting.scatter_matrix(df[df.columns[1:]], figsize=(6,6), color=colors, alpha=0.8, diagonal='kde')   #データのプロット
plt.show()
```

```
#import sklearn #機械学習のライブラリ
from sklearn.decomposition import PCA #主成分分析器
#主成分分析の実行
pca = PCA()
pca.fit(df.iloc[:, 1:])
```

```
PCA(copy=True, n_components=None, whiten=False)
```

```
# データを主成分空間に写像 = 次元圧縮
feature = pca.transform(df.iloc[:, 1:])
```

```
# 第一主成分と第二主成分でプロットする
plt.figure(figsize=(6, 6))
for x, y, name in zip(feature[:, 0], feature[:, 1], df.iloc[:, 0]):
    plt.text(x, y, name, alpha=0.8, size=10)
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8, color=colors)
plt.title("Principal Component Analysis")
plt.xlabel("The first principal component score")
plt.ylabel("The second principal component score")
plt.show()
```

```
import pylab as pl
from mpl_toolkits.mplot3d import Axes3D
fig = pl.figure(1)
ax = Axes3D(fig)
X = feature[:,0]
Y = feature[:,2]
Z = feature[:,3]
ax.scatter(X, Y, Z, c=colors)
ax.set_xlabel("sepal length")
ax.set_ylabel("petal length")
ax.set_zlabel("petal width")
pl.show()
```
